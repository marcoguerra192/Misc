# Group Testing

Let us try some tests for group sampling. As described in Phd Log on overleaf. <br>
Namely one finds that the condition for advantageous group sampling is
$ (1-p)^k > \frac{1}{k} $
which yields a $p_k = 1 - (1/k)^{1/k}$

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5000
k = np.array(range(1,N))
pLim = np.array( [ 1 - np.power( (1.0/x) , (1.0/x) ) for x in k ] )
```

```python
plt.plot(k,pLim)
```

A k-strategy costs, on average, $cN \left[ \frac{1}{k} + (1 - (1-p)^k) \right]$ whereas the individual strategy costs $cN$. Hence the k-strategy has a saving factor of $\frac{1}{k} + (1 - (1-p)^k)$. Let us plot it.

```python
p = np.linspace(0,1,200)
k1 = np.array( [ 1.0/1 + 1 - np.power((1- px),1) for px in p ] )
k2 = np.array( [ 1.0/2 + 1 - np.power((1- px),2) for px in p ] )
k3 = np.array( [ 1.0/3 + 1 - np.power((1- px),3) for px in p ] )
k20 = np.array( [ 1.0/20 + 1 - np.power((1- px),20) for px in p ] )
k100 = np.array( [ 1.0/100 + 1 - np.power((1- px),100) for px in p ] )
k1000 = np.array( [ 1.0/1000 + 1 - np.power((1- px),1000) for px in p ] )

limits = [ [0,1] ]
```

```python
plt.plot(p,k1,'g')
plt.plot(p,k2,'b')
plt.plot(p,k3,'r')
plt.plot(p,k20,'violet')
plt.plot(p,k100,'y')
plt.plot((0,1),(1,1), 'k--')


plt.legend(['k=1', 'k=2', 'k=3', 'k=20', 'k=100'], loc='upper left')
```

```python
def cost(k,p):
    return np.array( [ 1.0/k + 1 - np.power((1- px),k) for px in p ] )
```

```python
plt.figure(figsize=(10,8))
plt.plot(p,cost(2,p), 'r')
plt.plot(p,cost(3,p), 'b')
plt.plot(p,cost(4,p), 'g')
plt.plot(p,cost(5,p), 'violet')
plt.plot(p,cost(6,p), 'y')
plt.plot((0,1),(1,1), 'k--')
plt.xlabel('p')
plt.ylabel('Cost factor')

plt.legend(['k=2', 'k=3', 'k=4', 'k=5', 'k=6'], loc='upper left')
```
